```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cars_data = pd.read_csv('jacto/CarsData.csv')
```

```python
cars_data.head()
```

#MOSTRA A QUANTIDADE DE CARROS POR FABRICANTE

```python
manufacturer_count = cars_data['Manufacturer'].value_counts()
```

```python
plt.figure(figsize=(6, 3))
manufacturer_count.plot(kind='bar', color='skyblue')
plt.title('Carros por Fabricante')
plt.xlabel('Fabricante')
plt.ylabel('Contagem de Carros')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

#MOSTRA A MDIA DE MILHAS POR GALO POR FABRICANTE

```python
mpg_means = cars_data.groupby('Manufacturer')['mpg'].mean()
```

```python
plt.figure(figsize=(6, 3))
mpg_means.plot(kind='bar', color='skyblue')
plt.title('Mdia de MPG por Fabricante')
plt.xlabel('Fabricante')
plt.ylabel('Mdia de MPG')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
